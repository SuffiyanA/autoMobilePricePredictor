```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
pd.set_option('display.max_columns', None)
# fetch dataset 
automobile = fetch_ucirepo(id=10) 
  
# data (as pandas dataframes) 
X = automobile.data.features 
y = automobile.data.targets 
  
# metadata 
print(automobile.metadata) 
  
# variable information 
print(automobile.variables) 
```

```python
df = automobile.data['features']
df = df.drop(columns = 'normalized-losses')
df = df.dropna()
df
```

```python
df.info()
```

```python
df['fuel-system'].unique()
```

```python
FS_ranking_order = {'1bbl': 0, '2bbl': 1, 'spfi': 2, 'mpfi': 3, 'mfi': 4, 'spdi': 5, 'idi': 6}

#manual Label encoding to respect the hierarchy of fuel system complexity and price which can affect final price
def custom_encoder(label):
    return FS_ranking_order[label]

df['fuel-system_encoded'] = df['fuel-system'].apply(custom_encoder)
```

```python
df['engine-type'].unique()
```

```python
# Define the assumed price ranking order
ET_ranking_order = {'l': 0, 'ohcv': 1, 'ohc': 2, 'ohcf': 3, 'dohc': 4}

#manual Label encoding to respect the hierarchy of engine complexity and price which can affect final price

# Define a custom encoding function using the assumed price ranking order
def custom_encoder(label):
    return ET_ranking_order[label]

# Apply the custom encoding function to the 'engine' column
df['engine-type_encoded'] = df['engine-type'].apply(custom_encoder)
```

```python
df
```

```python
df['engine-location'].unique()
```

```python
#using One Hot Encoding for engine location since there are only two categories
df = pd.concat([df,pd.get_dummies(data = df['engine-location'] , prefix = 'eng-location' , dtype=float)], axis =1)
```

```python
#using One Hot Encoding for body-style since although the number of categories is large (5) the categories are not hierarchical
df = pd.concat([df,pd.get_dummies(data = df['body-style'] , prefix = 'body-style' , dtype=float)], axis =1)
```

```python
#using One Hot Encoding for fuel-type since there are only two categories
df = pd.concat([df,pd.get_dummies(data = df['fuel-type'] , prefix = 'fuel-type' , dtype=float)], axis =1)
```

```python
#using One Hot Encoding for 'make' since label encoding would create bias
df = pd.concat([df,pd.get_dummies(data = df['make'] , prefix = 'make' , dtype=float)], axis =1)
```

```python
#excluding columns of dtype "object"
dfLR=df.select_dtypes(include=None, exclude=object)
```

```python
dfLR
```

```python
dfLR.corr()
```

```python
X = dfLR.drop(columns='price')
y = dfLR['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Testing with LinearRegression

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
```

```python
predLR = LR.predict(X_test)
r2_score(y_test, predLR)
```

### Testing with ElasticNet Regression

```python
from sklearn.linear_model import ElasticNet
EN = ElasticNet(alpha=0.2)
EN.fit(X_train, y_train)
```

```python
predEN = EN.predict(X_test)
r2_score(y_test, predEN)
```

### Testing with Lasso Regression (L1)

```python
from sklearn.linear_model import Lasso
La = Lasso(alpha=0.9, max_iter=1500)
La.fit(X_train, y_train)
```

```python
predLa = La.predict(X_test)
r2_score(y_test, predLa)
```

## Result

### As evident by the results, Linear Regression with L1 Regularization (Lasso) gives the best R2 score and therefore we'll be using it in our application
